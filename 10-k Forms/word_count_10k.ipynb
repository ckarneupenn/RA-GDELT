This is code count the word frequency for following keywords in 10-k.

Data source:"/wrdsx/wrds/sec/wrds_clean_filings/"

Output: A tabular data with the word frequency and firms information mentioned these words

| Variable      | Description |
| ----------- | ----------- |
| firm_year      | Year of file filling time|
| firm_name   | Firm name        |
| firm_central_index      | Central Index Key of the firm|
| bits   | Frequency of words including "investment treaty","investment treaties","bilateral investment treaty","bilateral investment treaties","bit","bits"|
| pta   |Frequency of words including "trade agreement","trade agreements","pta","ptas","preferential trade agreement""preferential trade agreements"
| investment promotion  | Frequency of words including "investment promotion","investment promotions","promotion agency","promotion agencies"        |
| tax cut   | Frequency of words including "tax incentive","tax incentives", "tax break", "tax breaks","tax cut","tax cuts"     |

Todo: don't forget to change the file paths `input_path` & `output_path`

```python
import re
import pandas as pd
import pathlib
import numpy as np
```

```python
input_path="'/wrdsx/wrds/sec/wrds_clean_filings"
output_path="/home/mgmt/xiaoyiwu/wrds_word_count/data.csv"
```

```python
g1 = [
    "investment treaty",
    "investment treaties",
    "bilateral investment treaty",
    "bilateral investment treaties",
    "bit",
    "bits"]
p1 = re.compile("("+r"|".join((r"\b"+w+r"\b" for w in g1))+")")

g2 = [
    "trade agreement",
    "trade agreements",
    "pta",
    "ptas",
    "preferential trade agreement",
    "preferential trade agreements"]
p2 = re.compile("("+r"|".join((r"\b"+w+r"\b" for w in g2))+")")

g3 = [
    "investment promotion",
    "investment promotions",
    "promotion agency",
    "promotion agencies"]
p3 = re.compile("("+r"|".join((r"\b"+w+r"\b" for w in g3))+")")

g4 = [
    "tax incentive",
    "tax incentives",
     "tax break",
     "tax breaks",
     "tax cut",
     "tax cuts"]
p4 = re.compile("("+r"|".join((r"\b"+w+r"\b" for w in g4))+")")
```

```python
firm_name=[]
firm_central_index=[]
firm_year =[]
firm_file_name =[]
ls_bits=[]
ls_pta=[]
ls_ip=[]
ls_tax_cut=[]
re_type = re.compile(r'conformed submission type:	10-k')
re_type_2 = re.compile(r'form type:		10-k')
re_company_name = re.compile(r"company conformed name:			(.*)")
```

```python
files =  pathlib.Path(input_path).glob('**/*.txt')

for infile in files:

        file=open(infile).read().lower()
        if re_type.search(file) or re_type_2.search(file):
    
            firm_file_name.append(str(infile))

            tmp= re.search("company conformed name:			(.*)", file) 
            #if test_2<50:
              #test_2 +=1
            if tmp:
                firm_name.append(tmp.group(1))   
            else:
                firm_name.append(np.nan)
            
            tmp = re.search("filed as of date:		(.*)", file)
            if tmp:
                firm_year.append(tmp.group(1))
                #if test_4 < 50:    
                  #print("company year: ",tmp.group(1))
                  #test_4+=1
                  #print("test_4: ",test_4)
            else:
                firm_year.append(np.nan)
                
            tmp = re.search("CENTRAL INDEX KEY:			(.*)", file)
            if tmp:
                firm_central_index.append(tmp.group(1))
                  
            else: 
                firm_central_index.append(np.nan)
            
            a= p1.findall(file)

            b= p2.findall(file)
  
            c= p3.findall(file)
  
            d= p4.findall(file)
            
            if len(a)!=0:
              ls_bits.append(len(a))
            else:
              ls_bits.append(0)
              
            if len(b)!=0:
              ls_pta.append(len(b))
            else:
              ls_pta.append(0)
            
            if len(c)!=0:
              ls_ip.append(len(c))
            else:
              ls_ip.append(0)
            
            if len(d)!=0:
              ls_tax_cut.append(len(d))
            else:
              ls_tax_cut.append(0)
```

```python
print("finish!")
print(len(firm_name))
print(len(firm_year))
print(len(firm_central_index))
print(len(firm_file_name))
print(len(ls_bits))
print(len(ls_pta))
print(len(ls_ip))
print(len(ls_tax_cut))
```

```python
d = {"firm_name":firm_name, 
    "firm_central_index_key":firm_central_index, 
    "firm_year":firm_year, 
    "file name":firm_file_name,
    "bits":ls_bits,
    "pta":ls_pta,
    "investment promotion": ls_ip,
    "tax cut":ls_tax_cut,}
    
data = pd.DataFrame(d)

data
```

```python
data.to_csv(output_path)
```
